```python
billing_templates = [
    "I was charged twice for an in-game purchase",
    "Money was deducted but game coins were not credited",
    "Subscription payment succeeded but premium features are locked",
    "Refund for cancelled purchase has not been received",
    "Unexpected charges appeared on my GAME-ON account",
    "Payment failed but amount was deducted from my bank",
    "Billing amount is higher than what was shown in the app"
]

technical_templates = [
    "GAME-ON crashes every time I try to start a match",
    "The game freezes during gameplay and I lose progress",
    "Unable to connect to GAME-ON servers",
    "High ping and severe lag while playing online matches",
    "GAME-ON keeps showing network error despite stable internet",
    "Game updates fail to install properly",
    "Voice chat is not working during multiplayer games"
]
product_templates = [
     "In-game rewards are not credited after completing missions",
    "Purchased skins are not showing in my inventory",
    "Game characters are unbalanced and unfair",
    "Progress is not saved after finishing levels",
    "Achievements are not unlocking as expected",
    "Daily challenges are missing in GAME-ON",
    "Matchmaking is unfair and places me with very high-level players"
    "Weapons stats feel broken after the patch"
    "Rewards promised after the event were not given"
]
account_templates = [
    "Unable to log in to my GAME-ON account",
    "My account was banned without any explanation",
    "Password reset email is not arriving",
    "Lost access to my account after recent update",
    "Someone else accessed my GAME-ON account",
    "Account progress disappeared after logging in",
    "Unable to link my social media account to GAME-ON"
]

other_templates = [
    "Customer support is not responding to my tickets",
    "Very disappointed with the overall GAME-ON experience",
    "Game advertisements are too frequent and annoying",
    "Poor service and lack of proper support",
    "Need clarification about GAME-ON policies",
    "The app experience has worsened after recent updates",
    "I want to give feedback about the game"
]
```

```python
def add_noise(text):
    noise = [
        "please resolve asap",
        "this is unacceptable",
        "need urgent help",
        "happening for the last 3 days",
        "no response from support"
    ]
    return text + " " + random.choice(noise)
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

def generate_synthetic_complaints(n=1000):
    data = []
    
    categories = {
        "Billing": billing_templates,
        "Technical Support": technical_templates,
        "Product Issue": product_templates,
        "Account Management": account_templates,
        "Other": other_templates
    }
    
    channels = ["Email", "Chat", "Website", "Phone"]
    start_date = datetime(2024, 1, 1)

    for i in range(n):
        category = random.choice(list(categories.keys()))
        base_text = add_noise(random.choice(categories[category]))
        complaint_text = add_html_and_punctuation(base_text)

        submitted_via = random.choice(channels)
        created_date = start_date + timedelta(days=random.randint(0, 300))
        
        data.append({
            "complaint_id": 100000 + i,
            "complaint_text": complaint_text,
            "category": category,
            "submitted_via": submitted_via,
            "created_date": created_date.strftime("%d-%m-%Y")
        })

    return pd.DataFrame(data)
```

```python
import random

html_tags = [
    ("<p>", "</p>"),
    ("<b>", "</b>"),
    ("<i>", "</i>")
]

punctuation_noise = [
    "!",
    "??",
    "...",
    "!!",
    "##",
    ',',
    '.',
    '?'
    
]

def add_html_and_punctuation(text):
    # Add punctuation
    text = insert_random_punctuation(text)

    # Wrap with HTML
    if random.random() < 0.4:
        tag = random.choice(html_tags)
        text = f"{tag[0]}{text}{tag[1]}"

    # Add line break
    if random.random() < 0.3:
        text = text.replace(" ", "<br>", 1)

    return text
```

```python
import random

def insert_random_punctuation(text):
    words = text.split()
    
    # Only insert if sentence has enough words
    if len(words) < 4:
        return text
    
    # Probability control (e.g., 70%)
    if random.random() < 0.7:
        idx = random.randint(1, len(words) - 2)
        punct = random.choice(punctuation_noise)
        words[idx] = words[idx] + punct

    return " ".join(words)
```
